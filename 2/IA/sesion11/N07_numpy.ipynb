## NumPy

[NumPy](https://numpy.org/) es una librería de Python ampliamente utilizada para realizar cálculos numéricos en matrices y vectores multidimensionales. Proporciona un conjunto de funciones y herramientas poderosas para trabajar con datos numéricos, lo que la convierte en una opción popular entre científicos de datos, ingenieros y otros profesionales que realizan computación numérica en Python.







### Características principales

- **Matrices y arreglos multidimensionales:** NumPy ofrece un objeto de matriz multidimensional (`ndarray`) que permite almacenar datos de manera eficiente y realizar operaciones matemáticas en ellos.

- **Funciones matemáticas:** la librería proporciona una amplia gama de funciones matemáticas para realizar operaciones como suma, resta, multiplicación, división, exponenciación, trigonometría, y más, tanto a nivel de elemento como a nivel de matriz.

- **Indexación:** NumPy ofrece una sintaxis intuitiva para indexar, lo que permite acceder y manipular fácilmente datos específicos dentro de una matriz.

- **Operaciones de álgebra lineal:** NumPy incluye funciones para realizar operaciones de álgebra lineal, como la inversión de matrices, la factorización, la descomposición de valores singulares, entre otras.

- **Generación de números aleatorios:** NumPy proporciona funciones para generar números aleatorios de diferentes distribuciones, lo que resulta útil en simulaciones y experimentos numéricos.

- **Integración con otras bibliotecas:** NumPy se integra bien con otras bibliotecas populares de Python, como SciPy, Pandas, Scikit-Learn, Tensorflow, Matplotlib, etc., lo que permite construir aplicaciones complejas y realizar análisis de datos avanzados de manera eficiente.

### Filtrados, máscaras y selección de elementos

NumPy permite filtrar y seleccionar elementos de un vector que cumplan ciertas condiciones. Esto se puede hacer utilizando máscaras booleanas, que son vectores de valores `True` y `False` que indican si se cumple o no una condición.

````python
g = np.random.randint(1, 100, (8, 8))
display(g)
````

#### Ej. 9. Cree una máscara booleana para seleccionar los elementos de `g` que sean mayores que 50. Muestre la máscara y muestra el resultado de aplicar la máscara a `g`. Guarde el resultado en un nuevo vector `g_mayores_50`.

````python
# Mask de elementos de g mayores a 50
mask_mayores_50 = g > 50
display(mask_mayores_50)

# Elementos de g mayores a 50
g_mayores_50 = g[g > 50]
display(g_mayores_50)
````

### Estadísticas y operaciones de agregación

NumPy ofrece una amplia gama de funciones para realizar operaciones estadísticas y matriciales en vectores. Algunas de las operaciones más comunes incluyen la suma, la media, la mediana, la desviación estándar o la varianza de los elementos de un vector. Son útiles para resumir y analizar datos de manera eficiente.

#### Ej. 11. Calcule la suma, la media, la mediana, la desviación estándar y la varianza de los elementos del vector `g_mayores_50`. Explore la documentación de NumPy para encontrar las funciones adecuadas.

````python
# Suma de g_mayores_50
suma_g_mayores_50 = g_mayores_50.sum()
print(suma_g_mayores_50)

# Media de g_mayores_50
media_g_mayores_50 = g_mayores_50.mean()
print(media_g_mayores_50)

# Desviacon estandar de g_mayores_50
std_g_mayores_50 = g_mayores_50.std()
print(std_g_mayores_50)

#Varianza de g_mayores_50
var_g_mayores_50 = g_mayores_50.var()
print(var_g_mayores_50)
````

### Iteración

Aunque NumPy está optimizado para realizar operaciones vectorizadas en lugar de bucles, a veces es necesario iterar sobre los elementos de un vector. NumPy proporciona una forma eficiente de hacerlo utilizando la función `np.nditer()`, que permite recorrer los elementos de un vector de manera eficiente. Por ejemplo:

```py
for elem in np.nditer(vector):
    print(elem)
```

#### Ej. 12. Utilizando la función `np.nditer()`, itere simultáneamente sobre los elementos de los vectores `b` y `c` y muestre la raíz cuadrada del elemento más grande de cada par de elementos.

````python
# Crear el iterador de b y c
iterar_b_c = np.nditer([b, c])

#Iterar sobre los elementos de b y c
for b_i, c_i in iterar_b_c:

    max_element = max(b_i, c_i)
    sqrt_max_element = np.sqrt(max_element)
    print(sqrt_max_element)
````

### Optimalidad al vectorizar

Aunque lo anterior es un ejemplo de iteración, en general es preferible evitarla en NumPy y utilizar operaciones vectorizadas siempre que sea posible, ya que son más eficientes y fáciles de leer. A continuación se muestra una prueba empírica de la diferencia de rendimiento entre un bucle `for` y una operación vectorizada en NumPy. Se empleará el comando _magic_ `%%time` para medir el tiempo de ejecución de cada celda. Por ejemplo:

````python
%%time

# Bucle de ejemplo
for _ in range(1000000):
    pass  # No hace nada
````

#### Ej. 13. Cree un vector de `N` por `N` de elementos (`v_1m`) con números aleatorios enteros entre 1 y 1000 —primera celda—. Calcule la suma de los cuadrados de los elementos utilizando un bucle `for` —segunda celda— y una operación vectorizada en NumPy —tercera celda—. Mida el tiempo de ejecución de método y compare los resultados. Pruebe con `N` igual a 100, 1000, 10000..., y oberve cómo explota el tiempo de ejecución del bucle `for`.

````python
# Vector nxn de num aleatorios entre 1 y 1000
n = 10000
vector_nxn = np.random.randint(1, 1001, (n, n))
display(vector_nxn)
````

````python
%%time

# Calcular suma de los cuadrados de los elementos usando for
def suma_cuadrados(vector_nxn):
    suma = 0
    for i in range(n):
        for j in range(n):
            suma += vector_nxn[i, j] ** 2
    return suma
print(suma_cuadrados(vector_nxn))
````

````python
%%time

# Calcular suma de los cuadrados de los elementos usando vectores
suma_cuadrados_vector = (vector_nxn ** 2).sum()
print(suma_cuadrados_vector)
````
